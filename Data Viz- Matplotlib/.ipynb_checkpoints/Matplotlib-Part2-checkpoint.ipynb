## Working With Images
    - Read Image.
    - Slicing/Cropping Images.
    - Image Manipulation.
    - Saving Modified Images.

```python
# Matplotlib can deal with images cuz images can be represented as matrices and more precisely as numpy matrices.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image.
img=plt.imread('./pikachu.jpg')
```

```python
plt.imshow(img)
# plt.axis('off')
plt.grid(b=True)
plt.show()
```

```python
# (height,width,channel) of the image. 
# And its channel is 3 because it is a coloured image so it is formed using the primary colors: RGB.
img.shape
```

```python
# All the arrays in the img array has 3 values and they represent the intensity of each color in each pixel.
```

```python
# Now lets try to crop out pikachu face
# For this we will use the slicing of the img array
# img[rows,cols,channels]
pikachuFace=img[ 330:515 , 190:390 , :]
plt.imshow(pikachuFace)
plt.show()
```

```python
# Now lets try to remove Green color out of the image
# For that assign the second pixel of the array to 0. Since the color ider goes like: Red, Green, Blue
img_copy=img.copy()
img_copy[:,:,1]=0
plt.imshow(img_copy)
plt.show()
```

```python
# Saving the pikachu face image.
plt.imsave('./pikachu_face.jpg',pikachuFace)
```

```python
# Creating a 4 sided black border around the pikachu image
row=np.zeros((10,400,3),dtype='int') # Because we want black border
```

```python
plt.imshow(row)
plt.grid(b=False)
plt.show()
```

```python
# Now vertically stack the row on the top of the image and on the bottom of the image.
img_modified=np.vstack((row,img,row))
```

```python
plt.imshow(img_modified)
plt.grid(b=False)
plt.show()
```

```python
# Now similarly horizontally stack borders on the left and right of the image.
col=np.zeros((620,10,3),dtype='int')
```

```python
img_modified=np.hstack((col,img_modified,col))
```

```python
plt.imshow(img_modified)
plt.grid(b=False)
plt.show()
```

## Given the Movie DataSet
    - Visualize the length of each title against the frequency.
    - X-axis: Title length of each movie.
    - Y-axis: Frequency count--> Number of movies having the same number of characters.

```python
import pandas as pd
```

```python
movie=pd.read_csv('./movie_meta.csv')
```

```python
movie.columns
```

```python
movie['movie_title']
```

```python
# Each movie title has some spaces .. Remove those spaces and return the new length..
def preprocess(elem):
    return len(elem.strip())
```

```python
# Introduce a new column with this trimmed length for each movie_Title
movie['movie_title_len']=movie['movie_title'].apply(preprocess)
```

```python
plt.hist(movie['movie_title_len'],bins=50)
plt.xticks(range(0,100,5))
plt.xlabel("Movie Title Length")
plt.ylabel("Frequency")

plt.show()
```

```python
## Another way of doing it can be using the scatter plot.
# Create a dictionary with the frequency of each title length...
dic={}

for elem in movie['movie_title_len']:
    if dic.get(elem) is None:
        dic[elem]=1
    else:
        dic[elem]+=1
```

```python
x=dic.keys()
y=dic.values()
```

```python
plt.scatter(x,y)
plt.xticks(range(0,60,5))
plt.xlabel("Movie Title Length")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line Plots 

```python
x=np.arange(0,10)
y1= x**2
y2= 2*x+3
```

```python
plt.plot(x,y1)
plt.plot(x,y2)

plt.show() # Use this to show the graph that we have done so far
```

```python
# Different styles available in Matplotlib
themes=plt.style.available 
print(themes)
```

```python
# Use the plot theme.
plt.style.use('dark_background')
```

```python
# Line plot.
plt.plot(x,y1,color="red",marker='o',label='Iphone')
plt.plot(x,y2,color="green",marker='*',markersize='12',linestyle='dashed',label='Samsung')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price of smartphones over time")

plt.legend() # Used to include the labels in the graph.
plt.show()
```

## Scatter Plot
    - plt.scatter()
    - Scatter Data points across x & y axis

```python
plt.figure(figsize=(10,4)) # Adjust the size of the plot.

# Scatter Plot.
plt.scatter(x,y1,label="Iphone",color="red",marker="*")
plt.scatter(x,y2,label="Samsung",color="green")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price of smartphones over time")

plt.legend() # Used to include the labels in the graph.
plt.show()
```

```python
data=np.random.randn(100,2)
```

```python
# All rows and the first column on the x-axis.
# All rows and the second column on the y-axis.
plt.scatter(x=data[:,0],y=data[:,1],color="yellow")
plt.show()
```

## BarGraphs 
    - plt.bar()
    - plot a graph between bar and labels you provide

```python
x=np.arange(0,4)
car1=[40,17,28,12]
car2=[34,21,24,15]
```

```python
# Now we will use the bargraph to check the comparison between car1 and car2 sales over the span of 4 years.
plt.bar(x,car1,width=0.3,tick_label=[2018,2019,2020,2021],label='car1')
plt.bar(x+0.3,car2,width=0.3,label='car2')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.title('Sales comparison')
plt.xlim(-1,5) # To give padding on the x-axis. Similarly we can also give padding on the Y-axis.

plt.legend()
plt.show()
```

## Piechart 
    - plt.pie()
    - plot the pie chart with provided labels and values

```python
subjects=["Physics","Maths","Chemistry","English","Computer Science"]
marks=np.array([92,95,95,88,93])
```

```python
plt.pie(x=marks,labels=subjects)
plt.show()
```

```python
plt.pie(x=marks,labels=subjects,explode=[0,0.1,0,0,0],shadow=True,autopct="%0.1f")
plt.show()
```

## Histograms 
    - plt.hist()
    - Creating dummy data using Normal Distribution

```python
# Generates random data in Standard normal point.
x_snd=np.random.randn(500)
```

```python
# Mean=70, std=8.
Mathmean=70
Mathstd=8
Physicsmean=50
Physicsstd=10
mathsMarks=np.round(x_snd*Mathstd +Mathmean)
physicsMarks=np.round(x_snd*Physicsstd+Physicsmean)
```

```python
# Generated a Random marks of students with mean=70 and standard deviation=8.
# mathsMarks
```

```python
# bins is the width.
# binWidth= (maxValue-minValue)/(bins)
# Determine the intervals: x=minValue+binWindth, y=x+binWindth, y+binWidth,.....
# Then after that plt.hist() determines the intervals and plots the frequency of the elements in 
# intervals in the graph.
# alpha gives opacity to the graph.
plt.hist(mathsMarks,color='yellow')
plt.hist(physicsMarks,color='red',alpha=0.8)
plt.show()
```

## Subplots 
    - A way to create multiple plots[Graphs] in one figure

```python
x=np.arange(0,10,0.1)
y1= x**2
y2= x**3
y3= np.sin(x)
y4= np.cos(x)
```

```python
# plt.subplot(Rows,cols,fig_no)
plt.subplot(2,2,1)
plt.plot(x,y1,color='yellow')

plt.subplot(2,2,2)
plt.plot(x,y2,color='red')

plt.subplot(2,2,3)
plt.plot(x,y3,color='purple')

plt.subplot(2,2,4)
plt.plot(x,y4,color='orange')

plt.show()
```

## 3-D Graph 

```python
a=np.arange(4)
b=np.arange(4,7)
```

```python
a,b=np.meshgrid(a,b)
```

```python
a
```

```python
b
```

```python
# Making a 3-D axis for ploting a 3-d graph.
fig=plt.figure()
ax=fig.gca(projection='3d') # Returns an axis.
ax.plot_surface(a,b,a+b) # (Length, width, height) Height should be a numpy array.
plt.show()
```

```python
a=np.arange(0,5,0.2)
b=np.arange(5,10,0.4)
```

```python
a,b=np.meshgrid(a,b)
```

```python
fig=plt.figure()
ax=fig.gca(projection='3d')
ax.plot_surface(a,b,np.array([[1]]))
plt.show()
```
